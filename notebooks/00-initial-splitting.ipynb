```python
# load packages
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# load application dataset
applications = pd.read_csv("../datasets/raw/applications.csv")

# initial inspection
display(applications.info(), applications.head())
```

```python
# check column names
applications.columns
```

```python
count_missing = applications.isna().sum()
perc_missing = applications.isna().sum() * 100 / len(applications)

perc_missing_df = pd.DataFrame({
    "col_name": applications.columns,
    "count_missing":count_missing,
    "percent_missing": perc_missing
    })

perc_missing_df.sort_values(by="percent_missing", ascending=True, ignore_index=True, inplace=True)

perc_missing_df
```

```python
# rename column
applications = applications.rename(columns={"target":"credit_event"})

# remove sensitive variables
applications = applications.drop(columns=["code_gender", "age", "name_family_status"])

# display results
display(applications.shape, applications.head())
```

```python
# split 80% untuk train-test, 20% untuk validation
application_train_test_set, application_validation_set = train_test_split(applications, test_size=0.2, random_state=555)

# display results
display(application_train_test_set.shape, application_train_test_set.head(3), application_validation_set.shape, application_validation_set.head(3))
```

```python
# save datasets
application_train_test_set.to_csv("../datasets/raw/applications_train_test_set.csv", index=False)
application_validation_set.to_csv("../datasets/raw/applications_validation_set.csv", index=False)
```
